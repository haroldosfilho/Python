```python
!pip install pythonds
import math
from pythonds.basic.stack import Stack
import numpy
#NetworkX é um pacote Python utilizado para manipulação de grafos
import networkx as nx
import matplotlib.pyplot as plt 

from numpy import save
from numpy import load
#o módulo glob é usado para recuperar arquivos / nomes de caminhos que correspondam a um padrão especificado
import glob, os

def grafosnamemoria():
  print("\nGrafos na memória:")
  for file in glob.glob("*.npy"):
      print(file)

def matriztriangular(graf, n):
    for i in range(1, n):
        for j in range(i+1):
            graf[i, j] = graf[j, i]

def divideBy2(decNumber):
    remstack = Stack()

    while decNumber > 0:
        rem = decNumber % 2
        remstack.push(rem)
        decNumber = decNumber // 2

    binString = ""
    while not remstack.isEmpty():
        binString = binString + str(remstack.pop())
        
    return binString

def is_integer_num(n):
    if isinstance(n, int):
        return True
    if isinstance(n, float):
        return n.is_integer()
    return False

def getSum(n):  
     
    sum = 0
    for digit in str(n):   
      sum += int(digit)        
    return sum

def desenhagrafo(adjacent, N):
    G = nx.Graph() 
    for i in range(N): 
        for j in range(N): 
            if adjacent[i][j] == 1: 
                G.add_edge(i+1,j+1)  
    
    opts = { "with_labels": True, "node_color": 'y' }
    nx.draw(G, **opts)
    plt.show()        


def imprimematriz(graf, N):   
    for i in range(N):
        for j in range(N):
            print(graf[i][j], end = " ")
        print()
    print()

def grafoexistente(fileName):
  return os.path.isfile(fileName)

def elementosdamatrizsimetrica(graf, N, tipo):   
    # carrega elementos da matriz acima da diagonal principal
    if tipo=="a":
      for i in range(N):  # linhas
        for j in range(N):  # colunas
            if i<j:
              graf[i][j] = 1

    if tipo=="b":
       for i in range(N):  # linhas
        for j in range(N):  # colunas
            if (i<j and i==j-1):
              graf[i][j] = 1

    if tipo=="c":
       for i in range(N):  # linhas
        for j in range(N):  # colunas
            if (i<j and i==j-1) or (i==0 and j==N-1):
              graf[i][j] = 1
            
    if tipo=="d":
       for i in range(N):  # linhas
        for j in range(N):  # colunas
            if (i==0):
              graf[i][j] = 1

    if tipo=="e":
       for i in range(N):  # linhas
        for j in range(N):  # colunas
            if (i==0) or (i==j-1) or (i==1 and j==N-1):
              graf[i][j] = 1

    if tipo=="f":
      n1=int(input("Digite o número de vértices do primeira partição:\n"))
      n2=int(input("Digite o número de vértices do segunda partição:\n")) 
      for i in range(N):  # linhas
        for j in range(N):  # colunas
            if   (i<j and i<n1 and j>=n1):                   
              graf[i][j] = 1

    if tipo=="g":
       for i in range(N):  # linhas
        for j in range(N):  # colunas
            if (j<N/2 and i<N/2)and (i<j) and (getSum(divideBy2(j))-getSum(divideBy2(i))==1) and is_integer_num(numpy.log2(j-i)):                  
              graf[i][j] = 1
            elif (i<j) and (j>N/2 and i>=N/2) and (getSum(divideBy2(j))-getSum(divideBy2(i))==1) and is_integer_num(numpy.log2(j-i)):
              graf[i][j] = 1
            elif (i<j) and (i<N/2) and (j-i==N/2):
              graf [i][j]=1
            
             
              
               

        







def subMenucarregagrafo():
    grafosnamemoria()
    nomeGrafo = input("Informe o nome do grafo a ser carregado(sem extensão)!\n")
    data = load(nomeGrafo+'.npy')
    print(data)
    desenhagrafo(data, len(data))

def subMenuCriagrafo():

    numgrafo = 0
    while True:
      if numgrafo == 0:
          nomeGrafo = input("Informe nome do grafo(sem extensão)!\n")
          print("Informe a classe desejada\n")
          print("a=Completo Kn\n")
          print("b=Caminho Pn\n")
          print("c= Ciclo Cn\n")
          print("d=Estrela Sn, numero de vértices=n+1\n")
          print("e=Roda Wn, numero de vértices=n+1, \n")
          print("f=Bipartido completo Kn1,n1, numero de vértices=n1+n2\n")
          print("g=Cubo Qn, numero de vértices= 2ˆn \n")
          break
      else:
          nomeGrafo = input("Nome já foi usado. Informe um nome para o grafo(sem extensão)!\n")
      numgrafo = numgrafo + 1
      if not grafoexistente(nomeGrafo+'.npy'):
        break

    tipo = str(input("classe do grafo\n"))
    N = int(input("Digite o número de vértices:\n"))
    graf = numpy.zeros((N, N))
    print("Informe os elementos da matriz triangular superior da matriz de adjacência do Grafo:\n") 
    elementosdamatrizsimetrica(graf, N, tipo)
    matriztriangular(graf, N)     
    save(nomeGrafo+'.npy', graf)
    print("Matriz de Adjacência")
    for line in graf:
      print ('  '.join(map(str, line))) 
    print()   
    print()
    print()         
    desenhagrafo(graf, N)    

def menu():
    opcao=int(input('''Opções Disponíveis:\n1 - Carregar um grafo\n2 - Criar um grafo\n3 - Sair\nEscolha:\n '''))
    if opcao == 1:
        subMenucarregagrafo()
    elif opcao == 2:           
        subMenuCriagrafo()
    elif opcao == 3:
        exit()
    else:
        print("Opção errada, digite: 1,2 ou 3\n")
        menu()
    return opcao    

controlelaço = True
while (controlelaço):
    if menu() == 3:
        controlelaço = False         
```
